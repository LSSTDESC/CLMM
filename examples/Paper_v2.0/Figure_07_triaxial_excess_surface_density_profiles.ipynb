# To make a plot that compares the contribution to excess surface density profiles for weak lensing when the lens is triaxial

```python
import clmm
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
from default_config import dc2_cosmo as cosmo
from default_config import default_halo
```

```python
from paper_formating import add_grid, prep_plot
```

```python
# A sample Mass, concentration for a halo:
# prof_kwargs = dict(
#    r_proj=np.linspace(0.1, 5, 100),
#    mdelta=2.0e14,
#    cdelta=3.89,
#    z_cl=d0.47,
#    halo_profile_model="nfw",
#    delta_mdef=200,
#    massdef="critical",
#    cosmo=cosmo,
# )

prof_kwargs = dict(
    r_proj=np.linspace(0.1, 5, 100),
    cosmo=cosmo,
)
prof_kwargs.update(default_halo)

# Ellipticities:
ell_1_0 = 0.0  # q = 1.0, ellipticity is (1-q)/(1+q)
ell_0_7 = 0.176  # q = 0.7, ellipticity is (1-q)/(1+q)
ell_0_3 = 0.497  # q = 0.3, ellipticity is (1-q)/(1+q)
```

```python
"""
A Spherical Halo: (Axis ratio, q = 1.0 )
"""

ds_model_e_0 = clmm.compute_excess_surface_density_triaxial(
    **prof_kwargs,
    ell=ell_1_0,
    term="mono",
)
ds_const_model_e_0 = clmm.compute_excess_surface_density_triaxial(
    **prof_kwargs,
    ell=ell_1_0,
    term="quad_const",
)
ds_4theta_model_e_0 = clmm.compute_excess_surface_density_triaxial(
    **prof_kwargs,
    ell=ell_1_0,
    term="quad_4theta",
)

"""
An Elliptical Halo: (Axis ratio, q = 0.7)
"""
ds_model = clmm.compute_excess_surface_density_triaxial(
    **prof_kwargs,
    ell=ell_0_7,
    term="mono",
)
ds_const_model = clmm.compute_excess_surface_density_triaxial(
    **prof_kwargs,
    ell=ell_0_7,
    term="quad_const",
)
ds_4theta_model = clmm.compute_excess_surface_density_triaxial(
    **prof_kwargs,
    ell=ell_0_7,
    term="quad_4theta",
)


"""
An Elliptical Halo: (Axis ratio, q = 0.3)
"""
ds_model_0_3 = clmm.compute_excess_surface_density_triaxial(
    **prof_kwargs,
    ell=ell_0_3,
    term="mono",
)
ds_const_model_0_3 = clmm.compute_excess_surface_density_triaxial(
    **prof_kwargs,
    ell=ell_0_3,
    term="quad_const",
)
ds_4theta_model_0_3 = clmm.compute_excess_surface_density_triaxial(
    **prof_kwargs,
    ell=ell_0_3,
    term="quad_4theta",
)
```

```python
fig, ax = prep_plot(subplots=(1, 2), figsize=(18, 9))

sns.lineplot(
    x=prof_kwargs["r_proj"],
    y=ds_model_e_0 / 1e12,
    ax=ax[0],
    markers=True,
    ms=20.0,
    color="black",
    ls="dashdot",
    label="Spherical NFW",
)
sns.lineplot(
    x=prof_kwargs["r_proj"],
    y=ds_model / 1e12,
    ax=ax[0],
    color="darkorange",
    label="Elliptical NFW (q = 2/3)",
    ls="solid",
)
sns.lineplot(
    x=prof_kwargs["r_proj"],
    y=ds_model_0_3 / 1e12,
    ax=ax[0],
    color="darkorange",
    label="Elliptical NFW (q = 1/3)",
    ls="solid",
    alpha=0.4,
)
ax[0].set_yscale("log")
ax[0].set_xscale("log")
ax[0].set_xlabel("R [Mpc]", fontsize=11)
ax[0].set_ylabel(r"Monopole $\Delta\Sigma \,[M_{\odot} h/\rm{pc}^{2}]$", fontsize=11)
ax[0].tick_params(axis="both", which="both", labelsize=12)
ax[0].legend(fontsize=7)
plt.tight_layout()

axin1 = ax[0].inset_axes([0.18, 4, 1, 14], transform=ax[0].transData)
sns.lineplot(
    x=prof_kwargs["r_proj"],
    y=(1 - ds_model / ds_model_e_0) * 100,
    ax=axin1,
    color="crimson",
    label="q=2/3",
)
sns.lineplot(
    x=prof_kwargs["r_proj"],
    y=(1 - ds_model_0_3 / ds_model_e_0) * 100,
    ax=axin1,
    color="crimson",
    alpha=0.4,
    label="q=1/3",
)
axin1.set_xscale("log")
axin1.set_ylabel("Difference (%)", fontsize=10 / 18 * 11)
axin1.legend(fontsize=7)
# axin1.set_xlabel('r', fontsize=15)

axin1.tick_params(axis="both", labelsize=7)


sns.lineplot(
    x=prof_kwargs["r_proj"],
    y=ds_const_model_e_0 / 1e12,
    ax=ax[1],
    color="black",
    ls="dashdot",
    label="Spherical NFW",
)
sns.lineplot(
    x=prof_kwargs["r_proj"],
    y=ds_const_model / 1e12,
    ax=ax[1],
    color="crimson",
    label=r"Elliptical NFW (q = 2/3) | $4\theta$",
)
sns.lineplot(
    x=prof_kwargs["r_proj"],
    y=ds_4theta_model / 1e12,
    ax=ax[1],
    color="darkorange",
    label="Elliptical NFW (q = 2/3) | const",
    ls="dashed",
)

sns.lineplot(
    x=prof_kwargs["r_proj"],
    y=ds_const_model_0_3 / 1e12,
    ax=ax[1],
    color="crimson",
    label=r"Elliptical NFW (q = 1/3) | $4\theta$",
    alpha=0.4,
)
sns.lineplot(
    x=prof_kwargs["r_proj"],
    y=ds_4theta_model_0_3 / 1e12,
    ax=ax[1],
    color="darkorange",
    label="Elliptical NFW (q = 1/3) | const",
    ls="dashed",
    alpha=0.4,
)
# ax[1].set_yscale('log')
ax[1].set_xscale("log")
ax[1].set_xlabel("R [Mpc]", fontsize=11)
ax[1].set_ylabel(r"Quadrupole $\Delta\Sigma \,[M_{\odot} h/\rm{pc}^{2}]$", fontsize=11)
ax[1].tick_params(axis="both", which="both", labelsize=12)
ax[1].legend(fontsize=7)
plt.tight_layout()

add_grid(ax[0])
add_grid(ax[1])
add_grid(axin1)

plt.savefig("triaxiality.png")
```
